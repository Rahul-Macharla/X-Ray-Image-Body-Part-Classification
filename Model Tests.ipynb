```python
from keras.models import load_model
model = load_model('D:/DATA SCIENCE/DataSets/Research/Pixxort/Body Part detection/44right/model.h5')
#44right
```

```python
classes = { 0:'Humerus',
            1:'Pelvis', 
            2:'Shoulder',
            3:'Forearm',
            4:'Wrist',
            5:'Femur',
            6:'Knee',
            7:'Tibia',
            8:'Ankle',
            9:'Foot',
            }
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
def test_on_img(img):
    data=[]
    image = Image.open(img).convert('RGB')
    image = image.resize((30,30))
    data.append(np.array(image))
    X_test=np.array(data)
    Y_pred = model.predict_classes(X_test)
    return image,Y_pred
```

```python
plot,prediction = test_on_img(r'D:\DATA SCIENCE\DataSets\Research\Pixxort\Body Part detection\Test\1.jpg')
s = [str(i) for i in prediction] 
a = int("".join(s)) 
print("Predicted x-ray image is: ", classes[a])
plt.imshow(plot)
plt.show()
```
